```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')

# Read the CSV file into a Pandas DataFrame
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DARWIN.csv')

# Print the DataFrame
data
```

```python
data.dtypes
```

```python
#dropping ID which is unnecassary
data = data.drop('ID', axis=1)
```

```python
#factorize the 'class' column
data['class'] = pd.factorize(data['class'])[0]

#print the DataFrame
data
```

```python
#split feature and target

X = data.drop('class', axis=1)
y = data['class']
```

```python
#train test split

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
#import randomforest, adaboost
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
clf1 = RandomForestClassifier(n_estimators=10)
clf2 = AdaBoostClassifier(n_estimators=100, algorithm= 'SAMME', random_state=0)
```

```python
#fitting
clf1.fit(X_train, y_train)
clf2.fit(X_train, y_train)
```

```python
#prediction
clf1_pred = clf1.predict(X_test)
clf2_pred = clf2.predict(X_test)
```

```python
#accuracy of both matrics
from sklearn import metrics
print("Accuracy of RandomForest:",metrics.accuracy_score(y_test, clf1_pred))
print("Accuracy of AdaBoost:",metrics.accuracy_score(y_test, clf2_pred))
```

```python
#Best parameters CLF1
from sklearn.model_selection import GridSearchCV

param_grid1 = {
    'n_estimators': [50, 100, 150],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search1 = GridSearchCV(clf1, param_grid1, cv=5)
grid_search1.fit(X_train, y_train)

print("Best parameters:", grid_search1.best_params_)
print("Best score:", grid_search1.best_score_)
```

```python
#RF Prediction with best peram
clf1_best = RandomForestClassifier(**grid_search1.best_params_)
clf1_best.fit(X_train, y_train)
clf1_best_pred = clf1_best.predict(X_test)
print("Accuracy of RandomForest after tuning:",metrics.accuracy_score(y_test, clf1_best_pred))
```

```python
#Best parameters CLF1
param_grid2 = {
    'n_estimators': [50, 100, 150],
    'learning_rate': [0.5, 1.0, 1.5],
    'algorithm': ['SAMME', 'SAMME.R']
}

grid_search2 = GridSearchCV(clf2, param_grid2, cv=5)
grid_search2.fit(X_train, y_train)

print("Best parameters:", grid_search2.best_params_)
print("Best score:", grid_search2.best_score_)
```

```python
#Adaboost Prediction with best Param
clf2 = AdaBoostClassifier(n_estimators=150, learning_rate=1.0, algorithm='SAMME')
clf2.fit(X_train, y_train)
clf2_pred = clf2.predict(X_test)
print("Accuracy of AdaBoost after tuning:",metrics.accuracy_score(y_test, clf2_pred))
```

**STACKING**

```python
# Importing dt, knn, lr, nb
from sklearn.ensemble import StackingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB

# Define the base models
estimators = [
    ('dt', DecisionTreeClassifier()),
    ('knn', KNeighborsClassifier()),
    ('lr', LogisticRegression()),
    ('nb', GaussianNB()),
]

# Define the final estimator
final_estimator = grid_search1.best_estimator_

#create the StackingClassifier
clf = StackingClassifier(estimators=estimators, final_estimator=final_estimator)

#fit the StackingClassifier
clf.fit(X_train, y_train)

# evaluate the StackingClassifier on the test data
accuracy = clf.score(X_test, y_test)

#print accuracy
print("Accuracy of StackingClassifier:", accuracy)
```

**Voting Classifier**

```python
#importing voting classifier
from sklearn.ensemble import VotingClassifier

# Create a voting classifier with the best parameters for both clf1 and clf2
voting_clf = VotingClassifier(estimators=[('rf', grid_search1.best_estimator_), ('ada', grid_search2.best_estimator_)], voting='hard')

# fiting the voting classifier on the training data
voting_clf.fit(X_train, y_train)

# evaluate the voting classifier on the test data
accuracy = voting_clf.score(X_test, y_test)

#print accuracy
print("Accuracy of voting classifier:", accuracy)
```

Initially, the accuracy of the RandomForestClassifier was 0.88, and that of the AdaBoostClassifier was 0.85. However, after tuning, the accuracy of the RandomForestClassifier decreased to 0.82, while that of the AdaBoostClassifier increased to 0.88. The StackingClassifier, utilizing Decision Trees, K-Nearest Neighbors, Logistic Regression, and Naive Bayes, achieved an accuracy of 0.88. Similarly, the VotingClassifier also achieved an accuracy of 0.88.

**Strenght and Weakness:**

Ensemble methods combine multiple models to make better predictions. They're great because they improve accuracy by taking advantage of different model strengths and reducing errors.Ensemble techniques are versatile and can be used across different machine learning tasks like classification, regression.They employ diverse models such as decision trees, neural networks, or svm to enhance predictive performance. They are also able to handle mistakes better and work well with various types of data.

 However, they can be complex and take more time to train because they involve multiple models. Understanding why they make predictions can be hard, and sometimes they might still make mistakes if not set up properly.Adjusting hyperparameters for both individual models and ensemble methods is often a demanding task, necessitating experimentation due to its complexity. Plus, they rely on the quality of individual models, so if those are weak or similar, the ensemble won't perform well.

  Overall, ensembles are powerful tools for improving prediction accuracy, but they require careful setup and might not always be easy to understand.
